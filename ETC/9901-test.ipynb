```python
from Crypto.Cipher import AES
from Crypto.Random import get_random_bytes
from Crypto.Protocol.KDF import PBKDF2
from Crypto.Util.Padding import pad, unpad
import base64

# 이건 안 먹나???
dkLen=128
count=10000000000

def generate_key(password):
    # PBKDF2를 사용하여 키 생성
    salt = get_random_bytes(1024)
    key = PBKDF2(password, salt, dkLen=32, count=1000000)
    return key, salt

def encrypt(plaintext, password):
    key, salt = generate_key(password)
    cipher = AES.new(key, AES.MODE_CBC)
    ciphertext = cipher.encrypt(pad(plaintext.encode('utf-8'), AES.block_size))
    iv = base64.b64encode(cipher.iv).decode('utf-8')
    encrypted_text = base64.b64encode(ciphertext).decode('utf-8')
    return iv, salt, encrypted_text

def decrypt(iv, salt, encrypted_text, password):
    key = PBKDF2(password, salt, dkLen=32, count=1000000)
    cipher = AES.new(key, AES.MODE_CBC, base64.b64decode(iv))
    decrypted_text = unpad(cipher.decrypt(base64.b64decode(encrypted_text)), AES.block_size)
    return decrypted_text.decode('utf-8')

def main():
    # 암호화할 문자열 입력 받기
    plaintext = input("암호화할 문자열을 입력하세요: ")
    # 암호화할 패스워드 입력 받기
    password = input("암호화에 사용할 패스워드를 입력하세요: ")

    # 암호화
    iv, salt, encrypted_text = encrypt(plaintext, password)
    print("암호화된 문자열:", encrypted_text)

    # 복호화
    decrypted_text = decrypt(iv, salt, encrypted_text, password)
    print("복호화된 문자열:", decrypted_text)

    print("iv:", iv, "\nsalt:", salt, "\n")

# if __name__ == "__main__":
#    main()
```
